```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def confidence_interval_a(sigma, gamma, n):    
    X = np.random.normal(0, sigma, n)
    sumX = np.sum(X ** 2)
    l = sumX / stats.chi2.ppf((1 + gamma) / 2, n)
    r = sumX / stats.chi2.ppf((1 - gamma) / 2, n)
    return r - l
```

```python
def confidence_interval_b(sigma, gamma, n):    
    X = np.random.normal(0, sigma, n)
    sumX = np.mean(X) ** 2
    l = n * sumX / (stats.norm.ppf((3 + gamma) / 4) ** 2)
    r = n * sumX / (stats.norm.ppf((3 - gamma) / 4) ** 2)
    return r - l
```

```python
def plot(sigma, gamma, interval):
    N = np.arange(100, 10000, 100)    
    y = [np.mean([interval(sigma, gamma, n) for _ in range(1000)]) for n in N]
    plt.plot(N, y)
    plt.xlabel('Sample size')
    plt.ylabel('Interval length')
```

```python
plot(1, 0.5, confidence_interval_a)
```

```python
plot(1, 0.5, confidence_interval_b)
```
